```python
import pandas as pd
import numpy as np
import random
import scipy.stats as st
from matplotlib import pyplot as plot
data=pd.read_csv('operations.csv',parse_dates=[0])
data.head(5)
depense=data[data.amount<0]
correlation=st.pearsonr(depense['balance_bef_trn'],depense['amount'])
print(correlation[0])
cov=np.cov(depense['balance_bef_trn'],depense['amount'])
print(cov[1,0])
plot.scatter(depense['balance_bef_trn'],-depense['amount'],c='blue')
plot.xlabel("balance before transactio")
plot.ylabel('amount')
plot.title('amount=f(balance)')
plot.show()
depense[depense['categ']=='GROCERIES'].head(5)
```
