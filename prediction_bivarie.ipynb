```python
import pandas as pd
import numpy as np
import random
from matplotlib import pyplot as plot
data=pd.DataFrame(columns=['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8'])
print(data)
for i in range(8):
    q=[np.random.randint(2) for _ in range(1000)]
    data['Q'+str(i+1)]=q
number_repartition=[len(data[data['Q'+str(i+1)]==1]) for i in range(8)]
tab_reparition=pd.DataFrame(data.keys(),columns=['questions'])
tab_reparition['num_true_responses']=number_repartition
print(tab_reparition)
plot.bar(tab_reparition['questions'],tab_reparition['num_true_responses'])
plot.show()
note_finale=[data.iloc[i].sum() for i in range(1000)]
data['sexe']=[random.choice(["M","F"]) for _ in range(1000)]
data['final_note']=note_finale
num_men_with_average=len(data[(data['sexe']=="M") & (data['final_note']>4)])
num_women_with_average=len(data[(data['sexe']=="F") & (data['final_note']>4)])
person_with_average=num_men_with_average+num_women_with_average
num_men=len(data[data['sexe']=='M'])
num_women=len(data[data['sexe']=='F'])
plot.pie([num_men_with_average,num_women_with_average,num_men-num_men_with_average,num_women-num_women_with_average]
         ,labels=["men with average",'women with average','men_under_average','women_under_average'], autopct='%1.1f%%')
plot.show()
plot.pie([num_men_with_average,num_men-num_men_with_average]
         ,labels=["men with average",'men_under_average'], autopct='%1.1f%%')
plot.show()
plot.pie([num_women_with_average,num_women-num_women_with_average]
         ,labels=["women with average",'women_under_average'], autopct='%1.1f%%')
plot.show()

```
